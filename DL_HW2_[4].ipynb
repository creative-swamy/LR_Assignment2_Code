```
def plot_epochs(lc,name, offset = 2000):
  
  plt.figure(figsize=(10,10))
  plt.xlabel('epochs')
  plt.ylabel('PPL')
  lim = min(np.max(lc.item(0)['train_ppls']),np.max(lc.item(0)['val_ppls']))
  plt.ylim(0,offset)
  plt.xticks(range(0,41,5))
  plt.plot(lc.item(0)['train_ppls'], color = 'blue', label = 'training')
  plt.plot(lc.item(0)['val_ppls'], color = 'green', label = 'validation')
  plt.title('Learning curves - epochs for Experiment : %s' % name)
  plt.legend()
  plt.show()
```

```
def plot_clock(lc, name, time, offset = 2000):
  plt.figure(figsize=(10,10))
  plt.xlabel('wall clock times (s)')
  plt.ylabel('PPL')
  lim = min(np.max(lc.item(0)['train_ppls']),np.max(lc.item(0)['val_ppls']))
  plt.ylim(0,offset)
  times = int(time) * np.arange(0,40)
  plt.xticks(times[0::5])
  plt.plot(times, lc.item(0)['train_ppls'], color = 'blue', label = 'training')
  plt.plot(times, lc.item(0)['val_ppls'], color = 'green', label = 'validation')
  plt.title('Learning curves - wall-clock-time for Experiment : %s' % name)
  plt.show()
```

```
colors = ['green','blue','red','yellow','cyan','magenta','tab:purple', 'tab:brown', 'tab:pink']
```

```
def plot_all_epochs(lcs, labels, name,offset = 2000,category = 'Architecture'):
  
  plt.figure(figsize=(10,10))
  plt.xlabel('epochs')
  plt.ylabel('PPL')
#   lim = min(np.max(lc.item(0)['train_ppls']),np.max(lc.item(0)['val_ppls']))
  plt.ylim(0,offset)
  plt.xticks(range(0,41,5))
  
  for lc, label, color in zip(lcs,labels,colors):
    plt.plot(lc.item(0)['val_ppls'], color = color, label = label)

  plt.title('Validation curves - epochs for %s : %s' % (category,name))
  plt.legend()
  plt.show()
```

```
def plot_clock(lc, name, time, offset = 2000):
  plt.figure(figsize=(10,10))
  plt.xlabel('wall clock times (s)')
  plt.ylabel('PPL')
  lim = min(np.max(lc.item(0)['train_ppls']),np.max(lc.item(0)['val_ppls']))
  plt.ylim(0,offset)
  times = int(time) * np.arange(0,40)
  plt.xticks(times[0::5])
  plt.plot(times, lc.item(0)['train_ppls'], color = 'blue', label = 'training')
  plt.plot(times, lc.item(0)['val_ppls'], color = 'green', label = 'validation')
  plt.title('Learning curves - wall-clock-time for Experiment : %s' % name)
  plt.show()
```

```
def plot_all_clock(lcs, labels, times, name, offset = 2000,category = 'Architecture'):
  
  plt.figure(figsize=(10,10))
  plt.xlabel('wall clock time (s)')
  plt.ylabel('PPL')
#   lim = min(np.max(lc.item(0)['train_ppls']),np.max(lc.item(0)['val_ppls']))
  plt.ylim(0,offset)
  
  for lc, label, time, color in zip(lcs,labels, times, colors):
    clock_times = int(time) * np.arange(0,40)
#     print(time)
#     print(clock_times)
    plt.plot(clock_times, lc.item(0)['val_ppls'], color = color, label = label)
  
  plt.title('Validation curves - clock times for %s : %s' % (category,name))
  plt.legend()
  plt.show()
```
